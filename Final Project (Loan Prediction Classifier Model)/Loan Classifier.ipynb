```python
# This is the final project of the course
# We will create a Random Forest Classifier to predict Loan Approvals
# Importing Libraries
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
# Stored the of "loan-prediction-data" in object Loan_Data
Loan_Data = pd.read_excel("loan-prediction-data.xlsx")
Loan_Data.head()
```

```python
Loan_Data.info()
```

```python
Loan_Data.isnull().sum()
```

```python
Loan_Data["Credit_History"] = Loan_Data["Credit_History"].fillna(Loan_Data["Credit_History"].mode()[0])
Loan_Data["Self_Employed"] = Loan_Data["Self_Employed"].fillna(Loan_Data["Self_Employed"].mode()[0])
Loan_Data["Dependents"] = Loan_Data["Dependents"].fillna(Loan_Data["Dependents"].mode()[0])
Loan_Data["Gender"] = Loan_Data["Gender"].fillna(Loan_Data["Gender"].mode()[0])
Loan_Data["LoanAmount"] = Loan_Data["LoanAmount"].fillna(Loan_Data["LoanAmount"].mean())
Loan_Data.dropna(inplace=True)
# dipalying the frist 5 row 
Loan_Data.head()
```

```python
# checking the total null value from laon_Data file througth the isnull function 
Loan_Data.isnull().sum()
```

```python
Loan_Data.describe()
# cheackiing the information of data on the loan_Data
Loan_Data.info()
```

```python
LE_object = LabelEncoder()
Loan_Data["Gender"] = LE_object.fit_transform(Loan_Data["Gender"])
Loan_Data["Married"] = LE_object.fit_transform(Loan_Data["Married"])
Loan_Data["Self_Employed"] = LE_object.fit_transform(Loan_Data["Self_Employed"])
Loan_Data["Education"] = LE_object.fit_transform(Loan_Data["Education"])
```

```python
Loan_Data.head()
```

```python
Loan_Data["Property_Area"] = LE_object.fit_transform(Loan_Data["Property_Area"])
Loan_Data["Loan_Status"] = LE_object.fit_transform(Loan_Data["Loan_Status"])
Loan_Data.head()
```

```python
# Train-Test Split
x = Loan_Data.drop(["Loan_ID", "Loan_Status"], axis=1)
y = Loan_Data["Loan_Status"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=2)
print(x_train, "\n")
print(y_train, "\n")
print(x_test, "\n")
print(y_test, "\n")
```

```python
RFC_obj = RandomForestClassifier()
RFC_obj.fit(x_train, y_train)
```

```python
y_pred_known = RFC_obj.predict(x_train)
```

```python
# cheacking the accuracy of Loan prediction  classifier model 
metrics.accuracy_score(y_train, y_pred_known)
```

```python
y_pred_unknown = RFC_obj.predict(x_test)
```

```python
#accuracy_score
metrics.accuracy_score(y_test, y_pred_unknown)
```

```python
# confusion matrix
metrics.confusion_matrix(y_test, y_pred_unknown)
```
